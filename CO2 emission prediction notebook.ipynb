# Data Preprocessing pipeline

```python
x_train = df.drop(columns="CO2 Emissions(g/km)")
```

```python
x_train.head()
```

```python
y_train = df["CO2 Emissions(g/km)"]
```

```python
y_train.head()
```

# Classify into Numeric data & Categorical data

```python
numric_columns = x_train.select_dtypes(exclude = object).columns
```

```python
print(numric_columns)
```

```python
x_train.head()
```

```python
print('$'*100)
```

```python
categorical_column = x_train.select_dtypes(include = object).columns
print(categorical_column)
```

```python
numeric_feature = Pipeline(steps=[('handlingmissing',SimpleImputer(strategy='median')),
                                  ('scaling',StandardScaler(with_mean=False))])
print(numeric_feature)
```

```python
categorical_feature=Pipeline(steps=[('handlingmissing',SimpleImputer(strategy='most_frequent')),('encoding',OneHotEncoder()),
                                    ('scaling',StandardScaler(with_mean=False))])
print(categorical_feature)
```

```python
from sklearn.compose import ColumnTransformer
```

```python
processing = ColumnTransformer([('numric',numeric_feature,numric_columns ),
                                ('cat',categorical_feature,categorical_column)])
```

```python
processing
```

# Model Creation

```python
from sklearn.decomposition import TruncatedSVD
from sklearn.linear_model import LinearRegression
model1=Pipeline(steps = [('processing',processing),("pca",TruncatedSVD(n_components=210,random_state=0)),
                         ('modeling',LinearRegression())])
model1.fit(x_train,y_train)
```

```python
from sklearn.ensemble import AdaBoostRegressor, GradientBoostingRegressor
model2=Pipeline(steps = [('processing',processing),("pca",TruncatedSVD(n_components=210,random_state=0)),
                         ('modeling',AdaBoostRegressor())])
model2.fit(x_train,y_train)
```

```python
from sklearn.ensemble import AdaBoostRegressor, GradientBoostingRegressor
model3=Pipeline(steps = [('processing',processing),("pca",TruncatedSVD(n_components=210,random_state=0)),
                         ('modeling',GradientBoostingRegressor())])
model3.fit(x_train,y_train)
```

# Model validation and score computation

```python
print(model1.score(x_train,y_train))
print(model2.score(x_train,y_train))
print(model3.score(x_train,y_train))
```
